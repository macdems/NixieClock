```python
import time

import board
import digitalio
```

```python
(oa := digitalio.DigitalInOut(board.C0)).direction = digitalio.Direction.OUTPUT
(ob := digitalio.DigitalInOut(board.C1)).direction = digitalio.Direction.OUTPUT
(oc := digitalio.DigitalInOut(board.C2)).direction = digitalio.Direction.OUTPUT
(od := digitalio.DigitalInOut(board.C3)).direction = digitalio.Direction.OUTPUT
```

```python
for digit in range(10):
    oa.value = bool(digit & 0b0001)
    ob.value = bool(digit & 0b0010)
    oc.value = bool(digit & 0b0100)
    od.value = bool(digit & 0b1000)
    print(digit, f"{digit:04b}")
    time.sleep(1)

oa.value = ob.value = oc.value = od.value = False
```
