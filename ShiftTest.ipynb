```python
import time

import board
import digitalio
```

```python
(ser := digitalio.DigitalInOut(board.C0)).direction = digitalio.Direction.OUTPUT
(srclk := digitalio.DigitalInOut(board.C1)).direction = digitalio.Direction.OUTPUT
(rclk := digitalio.DigitalInOut(board.C2)).direction = digitalio.Direction.OUTPUT

rclk.value = False
srclk.value = False
```

```python
def write_byte(b):
    for i in range(8):
        ser.value = (b & 0x80) != 0
        srclk.value = True
        srclk.value = False
        b <<= 1
    ser.value = 0
    for i in range(8):
        srclk.value = True
        srclk.value = False
    rclk.value = True
    rclk.value = False
```

```python
for byte in range(256):
    write_byte(byte)
    # print(byte, f"{byte:08b}")
    time.sleep(0.1)
```

```python
while True:
    write_byte(0b01010101)
    time.sleep(1)
    write_byte(0b10101010)
    time.sleep(1)
```
